```python
import numpy as np

def f(x):
    return x**2


vf = np.vectorize(f)


def array_for(x):
    return np.array([f(xi) for xi in x])


def array_map(x):
    return np.array(map(f, x))


def fromiter(x):
    x = np.array(x)
    return np.fromiter((f(xi) for xi in x), x.dtype)


def vectorize(x):
    return  np.vectorize(f)(x)


def vectorize_without_init(x):
    return vf(x)



setup=lambda n: np.random.rand(n)
n_range=[2**k for k in range(15)]
kernels=[array_for, array_map, fromiter, vectorize, vectorize_without_init]

for call_f in kernels:
    call_f(n_range)
```
