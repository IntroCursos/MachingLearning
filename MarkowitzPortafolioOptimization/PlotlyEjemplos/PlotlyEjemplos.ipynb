```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
import plotly.tools as tls
```

```python
x=[-.7, .75]
y=[0, 0]
plt.scatter(x,y, marker='o', color="r")
plt.title('Quiver Plot with Points')
fig = plt.gcf()

plotly_fig = tls.mpl_to_plotly(fig)


x,y = np.meshgrid(np.arange(-2, 2, .2),
                  np.arange(-2, 2, .25))
z = x*np.exp(-x**2 - y**2)
v, u = np.gradient(z, .2, .2)

quiver_fig = FF.create_quiver(x, y, u, v,
                       scale=.25,
                       arrow_scale=.4,
                       name='quiver',
                       line=dict(width=1))

quiver_fig['data'].append(plotly_fig['data'][0])
plot_url = py.iplot(quiver_fig, filename='mpl-quiver-with-points')
```

```python
n = 50
x, y, z, s, ew = np.random.rand(5, n)
c, ec = np.random.rand(2, n, 4)
area_scale, width_scale = 500, 5

fig_puntos, ax = plt.subplots()
sc = ax.scatter(x, y, c=c,
                s=np.square(s)*area_scale,
                edgecolor=ec,
                linewidth=ew*width_scale)
ax.set_xlabel("Etiqueta x")
ax.set_ylabel("Etiqueta Y")
ax.set_title("El titulo")
ax.grid()
```

```python
py.iplot_mpl(fig_puntos)
```

```python
from plotly import tools
import plotly.plotly as py
import plotly.graph_objs as go
```

```python
trace1 = go.Scatter(x=[1, 2, 3], y=[4, 5, 6])
trace2 = go.Scatter(x=[20, 30, 40], y=[50, 60, 70])
trace3 = go.Scatter(x=[300, 400, 500], y=[600, 700, 800])
trace4 = go.Scatter(x=[4000, 5000, 6000], y=[7000, 8000, 9000])

fig = tools.make_subplots(rows=2, cols=2, subplot_titles=('Plot 1', 'Plot 2',
                                                          'Plot 3', 'Plot 4'))

fig.append_trace(trace1, 1, 1)
fig.append_trace(trace2, 1, 2)
fig.append_trace(trace3, 2, 1)
fig.append_trace(trace4, 2, 2)

fig['layout'].update(height=600, width=600, title='Multiple Subplots' +
                                                  ' with Titles')

iplot(fig, filename='make-subplots-multiple-with-titles')
```
