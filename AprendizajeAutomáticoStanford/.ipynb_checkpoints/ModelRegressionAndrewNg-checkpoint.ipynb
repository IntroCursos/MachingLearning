# Gradient Descent



```python
# Datos de entrada
x = np.arange(0,10) 

# Funcion hipotesis
h = lambda th0,th1,x: th0 + th1*x 

#Datos disque reales
HousePrice = h(1,0.5,x) + np.random.uniform(0,2,len(x)) 

# Grafica de datos
#plt.scatter(x,HousePrice,marker='x');
```

```python
# Algoritmo de gradient descent algorithm
ths = np.array([np.random.uniform(0,1),np.random.uniform(0,1)] )
alpha = .01
m = len(x)

for i in range(1000):
    for xi,yi in zip(x,HousePrice):
        th =ths[0]-(alpha/m) *np.sum( h(*ths,x=xi) - yi)
        ths[1] =ths[1]-(alpha/m) *np.sum(h(*ths,x=xi) - yi)*xi
        ths[0] = th
```

```python
import numpy as np

def f(x):
    return x**2


vf = np.vectorize(f)


def array_for(x):
    return np.array([f(xi) for xi in x])


def array_map(x):
    return np.array(map(f, x))


def fromiter(x):
    x = np.array(x)
    return np.fromiter((f(xi) for xi in x), x.dtype)


def vectorize(x):
    return  np.vectorize(f)(x)


def vectorize_without_init(x):
    return vf(x)



setup=lambda n: np.random.rand(n)
n_range=[2**k for k in range(15)]
kernels=[array_for, array_map, fromiter, vectorize, vectorize_without_init]

for call_f in kernels:
    call_f(n_range)
```
